```python
#Customer order list
#2.a-b
customer_orders.clear()
customer_product=input(f"available products are {products}:")
if customer_product not in products:
    customer_product2=input(f"product not available please select a product out of {products}:")
    if customer_product2 in products:
        customer_orders.add(customer_product2)
else:
    customer_orders.add(customer_product)

#2.c-d
add_product=str(input(f"would you like to add more products to your basket? (yes/no)"))
add_product=add_product.lower().strip()
while add_product=="yes":
        customer_product=input(f"available products are {products}:")
        if customer_product not in products:
            customer_product2=input(f"product not available please select a product out of {products}:")
            if customer_product2 in products:
                customer_orders.add(customer_product2)
        else:
            customer_orders.add(customer_product)
        add_product=input(f"would you like to add more products to your basket? (yes/no)")
        add_product=add_product.lower().strip()
print(customer_orders)
```

```python
#Total Products Ordered: The total number of products in the customer_orders set.
number_orders=len(customer_orders)
ordered_products=f"{int((number_orders/len(products))*100)}%"
order_status=(number_orders, ordered_products)
print(order_status)

print(f"Order Statistics:\nTotal Products Ordered: {number_orders}\nPercentage of Products Ordered: {ordered_products}")

#inventory update
CustOrderList=list(customer_orders)
for order in CustOrderList:
    inventory[order]=inventory[order]-CustOrderList.count(order)

print(inventory)
```
